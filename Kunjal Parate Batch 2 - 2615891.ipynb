# EDA Activity: Retail Sales Data Analysis


```python
# Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# LOAD DATASET

df = pd.read_csv("superstore.csv")

# Displaying first 5 rows
print(df.head())
```

# TASK 1: DATA LOADING & OVERVIEW

```python
# Question 1: Load the dataset
print("Dataset Loaded Successfully")

# Question 2: Shape of dataset
print("\nShape of Dataset:", df.shape)

print("Answer: The dataset contains",
      df.shape[0],
      "rows and",
      df.shape[1],
      "columns.")

# Question 3: Data types
print("\nData Types of Each Column:\n")
print(df.dtypes)

print("\nAnswer: Object columns are categorical and numeric columns are numerical.")

# Question 4: Identify categorical and numerical features

numerical_cols = df.select_dtypes(include=np.number).columns
categorical_cols = df.select_dtypes(include='object').columns

print("\nNumerical Features:")
print(numerical_cols)

print("\nCategorical Features:")
print(categorical_cols)
```


# TASK 2: DATA PROFILING


```python
# Question 1: Missing values
print("Missing Values:\n")
print(df.isnull().sum())

print("\nAnswer: The dataset contains very few or no missing values.")

# Question 2: Cardinality
print("\nCardinality of Categorical Features:\n")

for col in categorical_cols:
    print(col, ":", df[col].nunique())

print("\nAnswer: Product Name and Customer Name have high cardinality.")

# Question 3: Highest unique values

print("\nUnique Values Count:\n")
print(df.nunique().sort_values(ascending=False))

print("\nAnswer: Product Name, Order ID, and Customer Name have highest unique values.")
```

# TASK 3: MISSING VALUE ANALYSIS


```python
# Question 1
missing_values = df.isnull().sum()

print(missing_values)

print("\nAnswer: Missing values are not significant in this dataset.")

# Question 2

print("\nAnswer:")
print("Since missing values are very low, dropping them is acceptable.")
print("Imputation is needed only when missing values are large.")
```

# TASK 4: UNIVARIATE ANALYSIS


```python
# Question 1: Distribution of Sales

plt.figure(figsize=(10,6))

sns.histplot(df['Sales'], bins=30, kde=True)

plt.title("Distribution of Sales", fontsize=18)
plt.xlabel("Sales")
plt.ylabel("Frequency")

plt.grid(True)

plt.show()

print("Answer: Sales distribution is positively skewed.")
print("Most sales are small while few sales values are extremely high.")

# Question 2: Outliers in Profit

print("\nAnswer: Profit column is not available in this dataset.")
print("Therefore, profit outlier analysis cannot be performed.")

# Question 3: Most frequent category

plt.figure(figsize=(8,5))

sns.countplot(x='Category', data=df)

plt.title("Most Frequent Categories", fontsize=18)

plt.show()

print(df['Category'].value_counts())

print("\nAnswer: Office Supplies appears most frequently.")

```

# TASK 5: BIVARIATE ANALYSIS


```python
# Question 1

print("Answer: Profit column is unavailable.")
print("Hence Sales vs Profit analysis cannot be performed.")

# Question 2

print("\nAnswer: Average profit by category cannot be calculated because Profit column is unavailable.")

# Question 3: Best performing region

region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)

print("\nRegion-wise Sales:\n")
print(region_sales)

plt.figure(figsize=(8,5))

sns.barplot(x=region_sales.index, y=region_sales.values)

plt.title("Region-wise Sales Performance", fontsize=18)
plt.xlabel("Region")
plt.ylabel("Sales")

plt.show()

print("\nAnswer: The region with highest sales performs best.")

# Additional Visualization

plt.figure(figsize=(8,5))

sns.barplot(x='Category', y='Sales', data=df)

plt.title("Sales by Category", fontsize=18)

plt.show()
```

# TASK 6: PIVOT TABLE ANALYSIS


```python
# Question 1

print("Profit column is unavailable in this dataset.")
print("Therefore, Sales is used instead of Profit.\n")

# Create Pivot Table

pivot_sales = df.pivot_table(
    values='Sales',
    index='Category',
    columns='Region',
    aggfunc='sum'
)

print(pivot_sales)

# Heatmap Visualization

plt.figure(figsize=(10,6))

sns.heatmap(
    pivot_sales,
    annot=True,
    fmt='.0f',
    linewidths=1
)

plt.title("Sales by Category and Region", fontsize=18)

plt.show()

# Question 2

print("\nPatterns Observed:")
print("1. Some regions generate significantly higher sales.")
print("2. Technology and Office Supplies categories perform strongly.")
print("3. Certain category-region combinations contribute lower sales.")
```


# TASK 7: CORRELATION ANALYSIS


```python
# Correlation Matrix

corr_matrix = df.select_dtypes(include=np.number).corr()

print(corr_matrix)

# Heatmap

plt.figure(figsize=(8,5))

sns.heatmap(
    corr_matrix,
    annot=True,
    linewidths=1
)

plt.title("Correlation Matrix", fontsize=18)

plt.show()

# Question 1

print("Answer:")
print("Numerical columns show weak to moderate correlations.")
print("Sales does not strongly correlate with other numerical features.")

# Question 2

print("\nAnswer:")
print("Profit column is unavailable in this dataset.")
print("Therefore, Sales vs Profit correlation cannot be calculated.")
```

# TASK 8: PAIRPLOT EXPLORATION


```python
# Create copy
temp = df.copy()

# Convert Order Date properly

temp['Order Date'] = pd.to_datetime(
    temp['Order Date'],
    dayfirst=True
)

# Extract month

temp['Order Month'] = temp['Order Date'].dt.month

# Convert categorical columns into numeric codes

for col in ['Category', 'Region', 'Segment']:
    temp[col] = temp[col].astype('category').cat.codes

# Pairplot

sns.pairplot(
    temp[['Sales', 'Order Month', 'Category', 'Region']]
)

plt.show()

# Question 1

print("Answer:")
print("Sales varies across categories, regions, and months.")
print("Some months show higher sales values.")

# Question 2

print("\nAnswer:")
print("No strong clusters are visible.")
print("However, some high-sales records behave differently from the majority.")
```

# TASK 9: ANOMALY DETECTION


```python
# Question 1: Are there negative profits?

print("Answer:")
print("Profit column is not available in the dataset.")
print("Therefore, negative profit analysis cannot be performed.")

# Question 2: Are there extremely high sales values?

threshold = df['Sales'].quantile(0.95)

outliers = df[df['Sales'] > threshold]

print("\nExtremely High Sales Values:\n")

print(outliers[['Sales', 'Category', 'Region']].head())

print("\nAnswer:")
print("Yes, there are extremely high sales values in the dataset.")

# Boxplot Visualization

plt.figure(figsize=(10,5))

sns.boxplot(x=df['Sales'])

plt.title("Outlier Detection - Sales", fontsize=18)

plt.show()

# Question 3: What could be the reason?

print("\nPossible Reasons for High Sales Values:")
print("1. Bulk Orders")
print("2. Premium Products")
print("3. Corporate Purchases")
print("4. Seasonal Demand")
```
